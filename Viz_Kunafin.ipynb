## Массовые расстрелы в Соединенных Штатах Америки (1966-2017)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import squarify

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6

df = pd.read_csv('shooting.csv')
# Переведем столбец с датой в формат datetime
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")

#Добавим столбцы Год стрельбы
df['Year'] = pd.DatetimeIndex(df['Date']).year

df
```

```python
# Общее количество жертв по годам
total_victims  = df[['Total victims']+['Year']].groupby('Year').sum()
total_victims
```

```python
# Как видно из данных, начиная с 2015 года количество жертв в 2-3 раза превышает количество жертв за предыдущие годы. 
# Для корректной визуализации удалим из набора данных записи с инцидентами, старше 2014 года

large_values = df[df['Year'] > 2014].index
df.drop(large_values, inplace = True)

total_victims  = df[['Total victims']+['Year']].groupby('Year').sum()
total_victims
```

```python
# График с количеством жертв по всем годам
plt.plot(total_victims.index, total_victims)
plt.title('Массовые расстрелы')
plt.xlabel('Год')
plt.ylabel('Количество жертв, чел.') 
plt.legend(total_victims)
```

```python
# Как видо из графика, количество жертв с каждым годом растет, однако в период 2000-2007 гг. их количество не превышало 50 ч.
```

```python
# Приведем столбец с расой в порядок. NaN заменим на Unknown, удалим лишние пробелы, приведем в одинаковый регистр.
df['Race'] = df['Race'].replace(np.NaN,'Unknown')
df['Race'] = df['Race'].str.strip()
df['Race'] = df['Race'].str.capitalize()
# Количество жертв по расам
df['Race'].value_counts()
crimes_by_race = df.groupby(['Race'])['Total victims'].agg('sum')
crimes_by_race
```

```python
#Объединим расы с общими признаками в единые группы
df['Race'] = df['Race'].replace('Asian american','Asian')
df['Race'] = df['Race'].replace('Asian american/some other race','Asian')
df['Race'] = df['Race'].replace('Black american or african american','Black')
df['Race'] = df['Race'].replace('Black american or african american/unknown','Black')
df['Race'] = df['Race'].replace('Native american or alaska native','Native american')
df['Race'] = df['Race'].replace('Some other race','Other')
df['Race'] = df['Race'].replace('Unclear','Unknown')
df['Race'] = df['Race'].replace('White american or european american','White')
df['Race'] = df['Race'].replace('White american or european american/some other race','White')


df['Race'].value_counts()
```

```python
# Количество жертв по расам
crimes_by_race = df.groupby(['Race'])['Total victims'].agg('sum')
crimes_by_race
```

```python
crimes_by_race.plot(kind='bar', 
                    title='Расстрелы по расам',
                    xlabel='Раса',
                    ylabel='Количество жертв')
```

```python
# Как видно из диаграммы, больше всего совершают массовые расстрелы люди, относящиеся к расе, определенной как White.
```

```python
# Сколько людей погибает и получает ранения в год
fatal_injured  = df[['Fatalities']+['Injured']+['Year']].groupby('Year').sum()
fatal_injured.plot(kind='bar', 
                    title='Исходы (погибло/ранено)',
                    xlabel='Год',
                    ylabel='Количество людей')
```

```python
# В большинстве годов между количеством раненых и убитых разница небольшая, но можно заметить что в период с 
# 1989 по 1998 гг количество раненых в 1,5 - 2,5 превышало число убитых.
```

```python
# Вытащим из данных о местонахождении данные о наименованиях штатов, в которых произошли события.
```

```python
df['Location'].value_counts()
```

```python
# Последнее слово в строке является наименованием штата
df['State'] = df['Location'].str.split().str[-1]
df['State'].value_counts()
```

```python
# Приведем в порядок некоторые штаты
df['State'] = df['State'].replace('CA','California')
df['State'] = df['State'].replace('D.C.','Washington')
df['State'] = df['State'].replace('York','New York')


df['State'].value_counts()
```

```python
# Отберем 10 штатов, в котором наибольшее количество жертв массовых расстрелов.
victims_by_locations  = df[['Total victims']+['State']].groupby('State').sum().sort_values(by='Total victims', ascending=False)
victims_by_locations.head(10)
```

```python
# Построим treemap
squarify.plot(sizes=victims_by_locations['Total victims'].head(10), 
              label = victims_by_locations.index, 
              alpha=.5)
plt.axis('off')
plt.show()
```
