```python
import pandas as pd
df = pd.read_csv('pandas03/visit_log.csv', delimiter = ';')
df.head()

df.loc[(df.traffic_source == 'yandex') | (df.traffic_source == 'google'), 'source_type'] = 'organic'
df.loc[((df.traffic_source == 'paid') | (df.traffic_source == 'email')) & (df.region=='Russia'), 'source_type'] = 'ad'
df.loc[((df.traffic_source == 'paid') | (df.traffic_source == 'email')) & (df.region!='Russia'), 'source_type'] = 'other'
df.loc[df.source_type.isnull(), 'source_type']=df['traffic_source']

df['source_type'].value_counts()
#df
```

```python
import pandas as pd
df = pd.read_csv('pandas03/URLs.txt')
df.head(10)
# необязательная часть
def count_categories(row):
    start = row['url'].find('/',0)
    end = row['url'].find('/',1)
    while end != -1:
        return row['url'][start+1:end]
df['category'] = df.apply(count_categories, axis=1)
df['category'].value_counts()

# по заданию
df[df.url.str.contains('/[0-9]{8}-', regex=True)].head(10)
```

```python
import pandas as pd

df = pd.read_csv('ml-latest-small/ratings.csv', sep=',')


def diff_(df):
    return df.timestamp.max() - df.timestamp.min()

df['counts'] = df.groupby(['userId'])['rating'].transform('count')
df.loc[lambda row: row.counts >= 100].groupby('userId').apply(diff_).reset_index(name='AVG_lifetime_of_users')
```
