```python
import pandas as pd
df = pd.read_csv('ml-latest-small/ratings.csv', delimiter = ',')
df0 = pd.read_csv('ml-latest-small/movies.csv', delimiter = ',')
df_merge = df.merge(df0, on = 'movieId' )
def rating_class(row):
    if row['rating'] <= 2.0:
        return 'низкий рейтинг'
    elif row['rating'] <= 4.0:
        return 'средний рейтинг'
    elif (4.5 == row['rating']) or (5.0 == row['rating']) :
        return 'высокий рейтинг'
        return False

df_merge['class'] = df_merge.apply(rating_class, axis=1)
#df_merge['class'].value_counts()
df_merge
```

```python
import pandas as pd
df = pd.read_csv('ml-latest-small/keywords.csv', delimiter = ',')
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск','старый оскол'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
def city(row):
    for key, value in geo_data.items():
        for v in value:
            if v.lower() in row['keyword'].lower():
                return key
df['region'] = df.apply(city, axis=1).fillna('undefined')
df['region'].value_counts()
df
```
