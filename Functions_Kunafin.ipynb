```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
def document_find(num):
    found = False
    for doc in documents:
        if "number" in doc.keys() and doc["number"] == num:
            print(f'Владелец документа: ', doc["name"])
            found = True
    if not found:
            print("Документ не найден в базе")

def document_find(num):
    found = False
    for doc in documents:
        if "number" in doc.keys() and doc["number"] == num:
            print(f'Владелец документа: ', doc["name"])
            found = True
    if not found:
            print("Документ не найден в базе")

def directories_find(num):
    found = False
    for x in list(directories.items()):
        if num in x[1]:
            print(f'Документ хранится на полке: ', x[0])
            found = True
    if not found:
        print("Документ не найден в базе")

def full_info():
    for x in documents:
        for z in directories.items():
            if x['number'] in z[1]:
                print(f'№:',x['number']+',', 'тип: '+x['type']+',', 'владелец: '+x['name']+',','полка хранения:',z[0])

def directories_app(num):
    f = False
    for d in directories.items():
        if num in d[0]:
            print(f'Такая полка уже существует. Текущий перечень полок:', list(directories.keys()))
            f = True
    if not f:
        directories.setdefault(num,[])
        print(f'Полка добавлена. Текущий перечень полок:', list(directories.keys()))

def directories_rem(num):
    if num != None and directories.get(num) == []:
        del directories[num]
        return print(f'Полка удалена. Текущий перечень полок: {", ".join([k for k in list(directories.keys())])}.')                      
    elif num != None:
        return print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: \
{", ".join([k for k in list(directories.keys())])}.') 

def main(documents, directories):
    while True:
        com = input('Введите команду: ')
        if com == 'p':
            num = input("Введите номер документа: ")
            document_find(num)
        elif com == 's':
            num = input("Введите номер документа: ")
            directories_find(num)
        elif com == 'l':
            full_info()
        elif com == 'as':
            num = input("Введите номер полки: ")
            directories_app(num)
        elif com == 'ds':
            num = input("Введите номер полки: ")
            directories_rem(num)
        elif com == 'q':
            break
main(documents, directories)        
```
